Real Estate Market Insights: Exploratory Data Analysis of Zameen.com Listings

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_excel("scraped_zameen_data.xlsx")
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe(include='all')
```

```python
cols_needed = [
    "City",
    "Location",
    "Type",
    "Area",
    "Price",
    "Bedrooms",
    "Bathrooms",
    "Purpose",
    "Description"
]

df = df[cols_needed]
df.head()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df.replace(["-", " - ", "--", "â€”"], np.nan, inplace=True)
```

```python
import re

def convert_price(value):
    if pd.isna(value):
        return np.nan
    value = str(value).replace("PKR", "").replace("\n", " ").strip()
    if value in ["", "-", "--", "â€”"]:
        return np.nan
    num = float(re.findall(r"\d+\.?\d*", value)[0])
    if "Crore" in value:
        return num * 10000000
    if "Lakh" in value:
        return num * 100000
    if "Thousand" in value:
        return num * 1000
    return num

df["Price"] = df["Price"].apply(convert_price)
df["Price"].head()
```

```python

def convert_price(value):
    if pd.isna(value):
        return np.nan

    text = str(value).replace("PKR", "").replace("\n", " ").strip()

    # Handle missing or non-price text
    if text.lower() in ["", "-", "--", "â€”", "contact for price"]:
        return np.nan

    # Extract all numbers (handles ranges too)
    nums = re.findall(r"\d+\.?\d*", text)

    if not nums:
        return np.nan  # no numbers found

    # If range, pick the maximum instead of first 0
    num = float(nums[-1])

    # If number is 0, interpret as missing (optional but useful)
    if num == 0:
        return np.nan

    # Detect units
    if "crore" in text.lower():
        return num * 10000000
    if "lakh" in text.lower():
        return num * 100000
    if "thousand" in text.lower():
        return num * 1000

    return num
```

```python
def convert_area(v):
    if pd.isna(v):
        return np.nan
    v = str(v)
    if v in ["", "-", "--", "â€”"]:
        return np.nan
    num = float(re.findall(r"\d+\.?\d*", v)[0])
    if "Sq. Yd" in v:
        return num * 9
    if "Marla" in v:
        return num * 272.25
    if "Kanal" in v:
        return num * 5445
    # if already in sqft or plain number
    return num

df["Area_sqft"] = df["Area"].apply(convert_area)
df[["Area", "Area_sqft"]].head()
```

```python
df["Bedrooms"] = pd.to_numeric(df["Bedrooms"], errors="coerce")
df["Bathrooms"] = pd.to_numeric(df["Bathrooms"], errors="coerce")
```

missing values

```python
df.isna().sum()
```

```python
numeric_cols = ["Price", "Area_sqft", "Bedrooms", "Bathrooms"]

for col in numeric_cols:
    df[col] = df[col].astype(float)
    df[col] = df[col].fillna(df[col].median())
```

```python
df['City'].value_counts()
```

```python
cat_cols = ["City", "Location", "Type", "Purpose", "Description"]

for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isna().sum()
```

```python
valid_cities = ["Karachi", "Lahore", "Islamabad", "Rawalpindi",
                "Peshawar", "Quetta", "Multan", "Faisalabad",
                "Hyderabad", "Sialkot"]

def fix_city(x):
    if pd.isna(x):
        return x
    match, score = process.extractOne(str(x), valid_cities)
    return match if score >= 80 else x

df["City"] = df["City"].apply(fix_city)
df["City"].value_counts()
```

```python
valid_types = ["House", "Flat", "Plot", "Upper Portion", "Lower Portion"]

def fix_type(x):
    if pd.isna(x):
        return x
    match, score = process.extractOne(str(x), valid_types)
    return match if score >= 80 else x

df["Type"] = df["Type"].apply(fix_type)
df["Type"].value_counts()
```

```python
Q1 = df["Price"].quantile(0.25)
Q3 = df["Price"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["Price"] >= lower) & (df["Price"] <= upper)]
df.shape

```

```python
Q1_a = df["Area_sqft"].quantile(0.25)
Q3_a = df["Area_sqft"].quantile(0.75)
IQR_a = Q3_a - Q1_a

lower_a = Q1_a - 1.5 * IQR_a
upper_a = Q3_a + 1.5 * IQR_a

df = df[(df["Area_sqft"] >= lower_a) & (df["Area_sqft"] <= upper_a)]
df.shape
```

```python
df.loc[df["Area_sqft"] == 0, "Area_sqft"] = np.nan
```

```python
df['City'] = df['City'].str.strip()
df['City'] = df['City'].str.lower()
df['City'] = df['City'].str.replace(r'\s+', ' ', regex=True)
```

```python
df['City'] = df['City'].replace({
    'gujran wala': 'gujranwala',
    'gujran-wala': 'gujranwala',
    'gujranwala city': 'gujranwala'
})
```

```python
df['City'] = df['City'].str.title()
```

```python
df['City'].value_counts()
```

featured engineering

```python
df["Price_per_sqft"] = df["Price"] / df["Area_sqft"]
```

```python
median_price = df["Price"].median()
df["High_Price"] = (df["Price"] > median_price).astype(int)
```

```python
major_cities = ["Karachi", "Lahore", "Islamabad"]

df["City_Category"] = df["City"].apply(
    lambda x: "Major City" if x in major_cities else "Other City"
)

df[["City", "City_Category"]].head()
```
